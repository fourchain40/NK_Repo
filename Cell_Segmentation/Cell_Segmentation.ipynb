```python
relabeled,n_left = mh.labeled.filter_labeled(labeled, remove_bordering=True, max_size=10000)
imshow(relabeled)
show()
bboxes = mh.labeled.bbox(labeled)
for label, bbox in enumerate(bboxes):
    # Extract the bounding box coordinates
    min_row, min_col, max_row, max_col = bbox

    # Calculate the center point
    center_row = (min_row + max_row) / 2
    center_col = (min_col + max_col) / 2

    # Print the center coordinates of the current object
    print(f"Center of object {label}: ({center_row}, {center_col})")
```
